```python
# задача 8
def gen_posl(n):
    res_posl = []
    tek_num = 1
    count = 0
    while count < n:
        for _ in range(tek_num):
             if count < n:
                res_posl.append(tek_num)
                count += 1
        tek_num += 1
    return res_posl
n = int(input("Введите целое число    "))
posl = gen_posl(n)
print(*posl)
```
